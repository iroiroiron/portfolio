```python
%matplotlib inline
import numpy as np
import pandas as pd
import random
import copy
import os
import math
import datetime
import time
import operator
from numpy import sqrt
from tqdm.notebook import tqdm
from scipy import linalg
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics.cluster import adjusted_rand_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
#from sklearn.metrics import r2_score
import matplotlib.pyplot as plt

np.set_printoptions(precision=5, suppress=True)

#-------------------
class Data:
    def __init__(self,x,y,u,id):
        #説明変数
        self.x = x
        #目的変数
        self.y = y
        self.u = u #抽出クラスタかノイズクラスタかどうか
        self.id = id
        self.cluster_number = 0 #最終的なクラスタ番号
#-------------------        
        
    
```

```python
class GPCRM:
    #-------------------
    def __init__(self,X,Y,n_cluster,alpha,lam,labels,max_iter=1000,random_seed=50):
        self.X = X
        self.Y = Y     
        self.n_cluster = n_cluster
        self.xx = np.linspace(np.min(X)-1 , np.max(X)+1 , 500)
        self.alpha = alpha
        self.lam = lam
        self.labels = labels
        self.max_iter = max_iter
        self.random_seed = random_seed
        np.random.seed(random_seed)
    #-------------------
    
    #-------------------
    # gaussカーネル
    def gauss(self,x1,x2):
        return np.exp((-1 * self.alpha *(np.sum(x1 - x2)**2)) )
    #-------------------
    
    #-------------------
    # カーネル行列計算する関数
    def kernel_matrix(self,X):
        N = len(X)
        K = np.zeros((N , N))
        eta = 1e-6
        for i in range(N):
            for j in range(N):
                K[i , j] = self.gauss(X[i] , X[j])
        K = K + eta*np.eye(N)
        return K
    #-------------------
    
    #-------------------
    # k_✳︎の計算 
    def calc_k_a(self,X , x_bound):
        N = len(X)
        k_a = np.zeros(N)
        for k in range(N):
            k_a[k] = self.gauss(X[k] , x_bound )
        return k_a   
    #-------------------
    
    #-------------------
    def calc_label(self,u_list):
        b  = []
        for i in range(int(np.max(u_list))+1):
            b.append([idx for idx , value in enumerate(u_list) if value == i])
        d = []
        for i , value in enumerate(b):
            c = [0] * len(u_list)
            for k in value:
                c[k]  =  1
            d.append(c)
        d = np.array(d)
        return d
    #-------------------
    
    #-------------------
    def calc_U(self,exp_list):
        N = len(self.X)
        update_U = np.zeros(N)
        for k in range(N):
            error_list = []
            for i in range(self.n_cluster):
                error_list.append((self.Y[k] - exp_list[i][k]) ** 2)
            update_U[k] = np.argmin(error_list)
        return update_U
        #-------------------
    
    #-------------------
    def calc_objective_funcion(self,U,exp_list):
        u_list = self.calc_label(U)
        object_value = 0
        for i in range(self.n_cluster):
            for k in range(len(self.X)):
                object_value += np.round(u_list[i][k] * (self.Y[k] - exp_list[i][k])**2,4)
        return  object_value      
    #-------------------
    
    #-------------------
    def extract_cluster_data(self,i,U):
        u_list = []
        u_list.append([idx for idx,value in enumerate(U)if value == i])
        return self.X[u_list[0]],self.Y[u_list[0]]
    #-------------------
    
    #-------------------
    def gp_calc_exp_var(self,U):
        ex_exp_list = np.zeros([self.n_cluster,len(self.X)])
        ex_var_list = np.zeros([self.n_cluster,len(self.X)])
        
        for i in range(self.n_cluster):
            #帰属度に属するデータのみを集めたリスト
            c_x , c_y = self.extract_cluster_data(i,U)
            K = self.kernel_matrix(c_x)
            k_inv_ridge = np.linalg.inv(K + self.lam * np.eye(len(c_x)))
            #k_inv = np.linalg.inv(K)
            var_list = []
            for k,x in enumerate(self.X):
                k_a = self.calc_k_a(c_x,x)
                k_aa = self.gauss(x,x) + self.lam
                ex_exp_list[i,k] = k_a.T @ k_inv_ridge @ c_y
                ex_var_list[i,k] = k_aa - k_a.T @ k_inv_ridge @ k_a
        return ex_exp_list,ex_var_list
    #------------------
    
    #-------------------
    def gp_display_calc(self,U):
        ex_exp_list = np.zeros((self.n_cluster,len(self.xx)))
        ex_var_list = np.zeros((self.n_cluster,len(self.xx)))
        
        for i in range(self.n_cluster):
            #抽出クラスタに属するデータのみを集めたリスト
            c_x , c_y = self.extract_cluster_data(i,U)
            K = self.kernel_matrix(c_x)
            k_inv_ridge = np.linalg.inv(K + self.lam * np.eye(len(c_x)))
            #k_inv = np.linalg.inv(K)
            for k,x in enumerate(self.xx):
                k_a = self.calc_k_a(c_x,x)
                k_aa = self.gauss(x,x) + self.lam
                ex_exp_list[i,k] = k_a.T @ k_inv_ridge @ c_y
                ex_var_list[i,k] = k_aa - k_a.T @ k_inv_ridge @ k_a
        return ex_exp_list,ex_var_list
    #------------------
    
    #-------------------
    def print_info(self,model,time):
        print("クラスタ数:",model.n_cluster)
        print("ARIの値:",model.ARI)
        print("RMSEの値:",model.RMSE_list)
        print("MAEの値:",model.MAE_list)
        print("r2scoreの値:",model.r2_score_list)
        print("実行時間:",time) 
        print("")
    #-------------------
    
    #-------------------
    def log_save(self,path,model,time):
        d_today = datetime.date.today()
        d_now = datetime.datetime.now() 
        f = open(path +"log.txt", 'w')
        f.write("実験に使用した記録まとめ\n")
        f.write(str(d_now) + "\n")
        f.write("クラスタ数:" + str(model.n_cluster) + "\n")
        f.write("ARIの値:" + str(model.ARI) + "\n")
        f.write("RMSEの値:" + str(model.RMSE_list) + "\n")
        f.write("MAEの値:" + str(model.MAE_list) + "\n")
        f.write("R2score(決定係数)の値:" + str(model.r2_score_list) + "\n")
        f.write("実行時間:" + str(time) + "\n")
        f.write("random_seed値:" + str(model.random_seed) +"\n")
        f.write("カーネルパラメータα:" + str(model.alpha) + "\n")
        f.write("正則化パラメータλ:" + str(model.lam) +"\n")
        f.close()
    #-------------------
    
    #-------------------
    def calc_MAE(self):
        MAE_list = np.zeros([self.n_cluster])
        for i in range(self.n_cluster):
            y_true = []
            y_pred = []
            for k in range(len(self.X)):
                if self.ex_U[k] == i: 
                    y_true.append(self.Y[k])
                    y_pred.append(self.ex_exp_list[i][k])
            MAE_list[i] = np.sqrt(mean_absolute_error(y_true, y_pred))
        return MAE_list
    #-------------------
    
    #-------------------
    def calc_RMSE(self):
        RMSE_list = np.zeros([self.n_cluster])
        for i in range(self.n_cluster):
            y_true = []
            y_pred = []
            for k in range(len(self.X)):
                if self.ex_U[k] == i: 
                    y_true.append(self.Y[k])
                    y_pred.append(self.ex_exp_list[i][k])
            RMSE_list[i] = np.sqrt(mean_squared_error(y_true, y_pred))
        return RMSE_list
    #-------------------
    
    #-------------------
    def calc_r2_score(self):
        r2_score_list = np.zeros([self.n_cluster])
        for i in range(self.n_cluster):
            y_true = []
            y_pred = []
            for k in range(len(self.X)):
                if self.ex_U[k] == i: 
                    y_true.append(self.Y[k])
                    y_pred.append(self.ex_exp_list[i][k])
            r2_score_list[i] = self.r2_score(y_true, y_pred)
        return r2_score_list
    #-------------------
    
    def r2_score(self,y_true,y_pred):
        residual = 0
        var_sum = 0
        for k in range(len(y_true)):
            residual += y_true[k] - y_pred[k]
            var_sum += y_true[k] - np.mean(y_true)
        return 1 - (residual / var_sum)
    #-------------------
    def plot(self,model,path):
        N = self.X.shape[0]
        c = ["b","r","g","c","m","y","k","w"]
        markers = ["o","+","x","v","o", "^", "<", ">"]
        plt.figure(figsize=(5,5))
        plt.xlim(np.min(self.X)-1,np.max(self.X)+1,1)
        plt.ylim(np.min(self.Y)-1,np.max(self.Y)+1,1)
        plt.xlabel("x")
        plt.ylabel("y")
      
        #データ点の散布図のプロット．クラスタごとに色分けして出力する．
        plot_u_list = []
        for i in range(model.n_cluster):
            plot_u_list.append([idx for idx , value in enumerate(model.ex_U) if value == i])
        for i in range(model.n_cluster):
                plt.scatter(model.X[plot_u_list[i]],model.Y[plot_u_list[i]],marker=markers[i],s=35,c = c[i])

        #期待値のプロット．
        for i in range(model.n_cluster):
            plt.plot(model.xx, model.display_exp_list[i],c=c[i])
        
        #分散のプロット
        for i in range(model.n_cluster):
            plt.fill_between(model.xx ,model.display_exp_list[i]- 2 * sqrt(model.display_var_list[i]),model.display_exp_list[i] + 2 * sqrt(model.display_var_list[i]),alpha = 0.5,color=c[i])
        
        plt.savefig(os.path.join(path)+"figure.png")
        plt.show()
    #-------------------
    
    #-------------------
    def fit(self):
        
        N = len(self.X)
        old_object_value = np.inf
        is_convergence = False
        U = np.zeros([N])
        for k in range(N):
            U[k] = np.random.randint(self.n_cluster)

        #収束するまで続ける
        while is_convergence == False:
            exp_list,var_list = self.gp_calc_exp_var(U)
            pre_U = U
            U = self.calc_U(exp_list)
            if all(pre_U == U):
                is_convergence = True
                object_value = self.calc_objective_funcion(U,exp_list)

        if old_object_value > object_value:
            self.ex_U = U
            old_object_value = object_value
            self.ex_exp_list = exp_list
            self.ex_var_list = var_list
            self.display_exp_list,self.display_var_list= self.gp_display_calc(U)

        
        self.ARI = adjusted_rand_score(self.labels,self.ex_U)
        self.RMSE_list = self.calc_RMSE()
        self.MAE_list = self.calc_MAE()
        self.r2_score_list = self.calc_r2_score()
    #-------------------
```

```python
#-------------------
def make_path(model,main_path):
    folder_name = "ARI" + str(round(model.ARI ,4)) + "ラムダ" + str(round(model.lam,4)) + "アルファ" + str(round(model.alpha,4))
    my_makedirs(main_path + "/" + folder_name)
    return main_path + "/" + folder_name + "/"
#-------------------

#-------------------
def my_makedirs(path):
    if not os.path.isdir(path):
        os.makedirs(path)
#-------------------

#-------------------

def plot(x,y,x_name,y_name,path):
    plt.figure(figsize=(5,5))
    plt.plot(x,y)
    plt.xlabel(x_name)
    plt.ylabel(y_name)
    plt.tight_layout()
    plt.savefig(os.path.join(path) + x_name +  "-" + y_name +".png")
    plt.show()
#-------------------

#-------------------
def main():
    t_all = time.time()
    
    alpha_list=[0.1,1,3,5]
    lam_list=[0.1]
    
    data_name = "wine_data"
    my_makedirs("./GPCRM_result/" + data_name )
    my_makedirs("./GPCRM_result/" + data_name + "/" + str(datetime.date.today()))
    main_path = "./GPCRM_result/" + data_name + "/" + str(datetime.date.today())
    
    data = np.loadtxt("./data/"+ data_name +"/data.txt", dtype=float)
    # X = data[0,:]
    # Y = data[1,:]
    #wine_dataの場合はこっち
    X = data[:,0]
    Y = data[:,1]
    labels = np.loadtxt("./data/"+ data_name +"/labels.txt")
    labels = np.array(labels)
    ARI_list = []
    ex_RMSE_list = []
    n_cluster = 3
    
    for i in tqdm(range(len(alpha_list))):
        for j in range(len(lam_list)):
            t1 = time.time()
            model = GPCRM(X,Y,n_cluster,alpha_list[i],lam_list[j],labels)
            model.fit()
            t2 = time.time()
            path = make_path(model,main_path)
            model.plot(model,path)
            model.print_info(model,t2-t1)
            ARI_list.append(model.ARI)
            ex_RMSE_list.append(model.RMSE_list)
            model.log_save(path,model,t2-t1)
    
    if len(lam_list) == 1:
        my_makedirs(main_path + "/" + "alpha"+str(alpha_list[0]) + "-" + str(alpha_list[len(alpha_list)-1]) + "lam" + str(lam_list[0]))
        path = main_path + "/" + "alpha"+str(alpha_list[0]) + "-" + str(alpha_list[len(alpha_list)-1]) + "lam" + str(lam_list[0]) + "/"
        plot(alpha_list,ARI_list,"Alpha","ARI",path)
    if len(alpha_list) == 1:
        my_makedirs(main_path + "/" + "lam"+str(lam_list[0]) + "-" + str(lam_list[len(alpha_list)-1]) + "alpha" + str(alpha_list[0]))
        path = main_path + "/" + "lam"+str(lam_list[0]) + "-" + str(lam_list[len(alpha_list)-1]) + "alpha" + str(alpha_list[0]) + "/"
        plot(lam_list,ARI_list,"Lambda","ARI",path)
        
#-------------------

#-------------------
if __name__ == "__main__":
    main()
#-------------------
```
