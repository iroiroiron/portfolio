```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

#-------------------
def make_path(data_name,ARI_value,n_cluster,seed):
    d_today = str(datetime.date.today())
    folder_name =  "ARI"+ str(ARI_value) + "シード値" + str(seed) + "クラスタ数" + str(n_cluster)
    path = "./CRM/"
    my_makedirs(path + d_today)
    my_makedirs(path + d_today + "/" + data_name)
    my_makedirs(path + d_today + "/" + data_name + "/" + folder_name)
    return path + d_today + "/" + data_name + "/" + folder_name + "/"
#-------------------

#-------------------
def my_makedirs(path):
    if not os.path.isdir(path):
        os.makedirs(path)
#-------------------


data_name = "data16"
data = np.loadtxt("./data/"+ data_name +"/data.txt", dtype=float)
X = data[0, :]
Y = data[1, :]
X = X.reshape(len(X),1)
Y = Y.reshape(len(Y),1)
labels = np.loadtxt("data/"+ data_name +"/labels.txt")
labels = np.array(labels)

fuzzy_value = 2
lam = 0.5
beta = 1
num_of_cluster = 2
N = X.shape[0]
U = []
beta_list = []
ARI_list = []
RMSE_list = []
MAE_list = []
r2_score_list = []

#ランダムな帰属度の初期値を作成
ex = []
ex_U = np.zeros([num_of_cluster,X.shape[0]])
seed = 34
random_list = [1]
#random_list = np.arange(1,10,1)
for k in range(N):
    U.append(np.random.randint(num_of_cluster))
for i in range(num_of_cluster):
    ex.append([idx for idx, value in enumerate(U)if value ==1])
for i in range(num_of_cluster):
    for j in ex[i]:
        ex_U[i][j] = 1

U = ex_U
num_of_data = X.shape[0]
for k in range(len(random_list)):  
    np.random.seed(random_list[k])
    t1 = time.time()
    for i in tqdm(range(300)):#試行回数300
        #step 1
        #クラスタの数だけ対角行列Uを作成
        diag_U_list = calc_diagonal_U_list(U)
        #step 2 
        beta_list = calc_beta_list(X,Y,diag_U_list)
        #step 3
        updated_U = calc_U(X,Y,beta_list)

        U = updated_U

        objective_value = calc_objective_function(X,Y,U,beta_list)
    t2 = time.time()
    ex_U = calc_U(X,Y,beta_list)
    predicted_labels = calc_U_label(ex_U)
    print(predicted_labels)
    ARI = adjusted_rand_score(labels,predicted_labels) #true,pred
    RMSE = calc_RMSE(X,Y,beta_list,predicted_labels)
    MAE = calc_MAE(X,Y,beta_list,predicted_labels)
    r2_score = calc_r2_score(X,Y,beta_list,predicted_labels)
    path = make_path(data_name,ARI,num_of_cluster,random_list[k])
    print_info(num_of_cluster,ARI,RMSE,MAE,r2_score,t2-t1)
    log_save(num_of_cluster,ARI,RMSE,MAE,r2_score,k,path,t2-t1)
    plt_FCR(X,Y,beta_list,path,predicted_labels)
```
