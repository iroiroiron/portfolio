```python
# 線形カーネル
def linear_kernel(x1,x2):
    return x1*x2
#周期カーネル
def kperiodic(params):
    [tau , sigma] = params
    return lambda x,y : exp(tau * cos((x - y) / sigma))

# マターン3
def kmatern3(x1,x2,sigma):
    return lambda x,y : let(abs(x-y) , lambda r : (1 + sqrt(3)*r / sigma) * exp(-sqrt(3) * r) /sigma)

# マターン5
def matern5(x1 , x2 , sigma):
    r = np.abs(x1-x2)
    return (1 + (np.sqrt(5)*r / sigma) + (5*(r**2)) /(3*(sigma**2))) * np.exp(-(np.sqrt(5) * r) /sigma)

# 指数カーネル
def exp_kernel(x1,  x2,  sigma):    
    return np.exp(-np.abs(x1 - x2) / sigma)
#クロネッカーのデルタ関数
def fdelta(x, a):
    if x == a:
        return 1.
    else:
        return 0.
# # RBFカーネル
# # パラメータ１つ
# def RBF_kernel(x1, x2, theta):
#     return  np.exp(- theta * (x1 - x2)**2)
# # パラメータ2つ
# def RBF_kernel(x1,x2,theta):
#     theta1 , theta2  = params
#     return theta1 * np.exp(-((x1 - x2)**2) / theta2)
# # パラメータ3つ
def RBF_kernel(x1, x2, theta):
    theta1 , theta2 , theta3 = theta
    return  theta1 * np.exp(- ((x1 - x2)**2) / theta2) + theta3 * fdelta(x1,x2)
def RBF_kernel_new(x1, x2, theta):
    tau , sigma , eta = theta
    return  np.exp(tau) * np.exp(- ((x1 - x2)**2) / np.exp(sigma)) + np.exp(eta) * fdelta(x1,x2)
def rbf(x1, x2 , beta):
    return np.exp(- beta * np.sum((x1 - x2))**2)
def rbf_saiteki(x1, x2 , beta):
#     print(beta)
    return np.exp(- 1 * beta * np.sum((x1 - x2))**2)


# 対称行列か判定する関数 
def is_symmetric(a):
    return np.array_equal(a, a.T)

# カーネル行列計算する関数
def kernel_matrix(xx , kernel , params):
    N = len(xx)
    K = np.zeros((N , N))
    eta = 1e-6
    for i in range(N):
        for j in range(N):
            K[i , j] = kernel(xx[i] , xx[j], params)
    K = K + eta*np.eye(N)
    return K

# k_✳︎の計算 
def calc_k_a(x_train , x_bound , kernel , params):
    N = len(x_train)
    k_a = np.zeros(N)
    for k in range(N):
        k_a[k] = kernel(x_train[k] , x_bound , params)
    return k_a    


def calc_label(u_list , num_of_cluster):
    b  = []
#     u_list = list(u_list)
    for i in range(int(np.max(u_list))+1):
        b.append([idx for idx , value in enumerate(u_list) if value == i])
    d = []
    for i , value in enumerate(b):
        c = [0] * len(u_list)
        for k in value:
            c[k]  =  1
        d.append(c)
    d = np.array(d)
    return d

    
#ーーーーーーーーーーーーーーーーーーーーーーーーーーーーーーーーーーーーー
def calc_U(X, Y , num_of_cluster , mu_list):
    N = len(X)
    update_U = np.zeros(N)
    for k in range(N):
        error_list = []
        for i in range(num_of_cluster):
            error_list.append((Y[k] - mu_list[i][k]) ** 2)
        update_U[k] = np.argmin(error_list)
    return update_U

# x_train , y_train , x_bound , kernel , params
def gp_c_regression(x_train , y_train , x_bound , kernel , params , num_of_cluster , U):
    N = len(x_train)

    result_mu_list = []
    result_var_list = []
    return_para = []

    for i in range(num_of_cluster):
        u_list = []
        u_list.append([idx for idx , value in enumerate(U) if value == i])

        c_x = x_train[u_list[0]]
        c_y = y_train[u_list[0]]
        
        params = optimize(params , c_x, c_y, kernel , init=params)
#         print("-最適化後のパラメータ-クラスタ", str(i) , params)
        return_para.append(params)
        
        K = kernel_matrix(c_x , kernel , params)
        K_inv = np.linalg.inv(K)
        mu_list = []
        var_list = []
        for x in x_bound:
            k_a=  calc_k_a(c_x , x , kernel ,params)
            k_aa = kernel(x , x , params)
            mu_list.append(k_a.T @ K_inv @ c_y)
            var_list.append(k_aa - k_a.T @ K_inv @ k_a)
        result_mu_list.append(mu_list)
        result_var_list.append(var_list)
        
#         old_U = U
#         U = calc_U(x_train , y_train , num_of_cluster , result_mu_list)
        
    return result_mu_list , result_var_list , return_para

# x_train , y_train , x_bound , kernel , params , num_of_cluster , U , _iter
def gpcrm_plot(x_train , y_train, x_bound , kernel , params , num_of_cluster , U , _iter, seed):
    
    xx = np.linspace(np.min(x_train) , np.max(x_train) , 500)
    object_value_count = 0
    object_value = 0
    old_object_value = 0
    
    for l in range(_iter):
        
        mu_pre , var_pre , para1 = gp_c_regression(x_train , y_train, x_bound , kernel , params , num_of_cluster , U)
        
        new_U = calc_U(x_train , y_train , num_of_cluster , mu_pre)
        old_U = U
        U = new_U
        
        # 分散の可視化のための部分
        mu_pre_bound , var_pre_bound , para2 = gp_c_regression(x_train , y_train, xx , kernel , params , num_of_cluster , U)

       # 収束条件追加
#         print("#" , object_value)

        hantei0 = np.array(old_U)
        hantei1 = np.array(U)
        if all(hantei1 == hantei0):
            object_value = calc_objective_function(x_train, y_train, U , mu_pre , num_of_cluster)
            print(object_value) 
            
            plot_u_list = []
            for i in range(num_of_cluster):
                plot_u_list.append([idx for idx , value in enumerate(old_U) if value == i])

            fig = plt.figure(figsize=(10,10))
            ax = fig.add_subplot(1,1,1)

            ax.scatter(x_train[plot_u_list[0]] , y_train[plot_u_list[0]] , s = 50 , c = "b" , marker = "$0$" , label = "cluster0")
            ax.scatter(x_train[plot_u_list[1]] , y_train[plot_u_list[1]] , s = 50 , c = "r" , marker = "$1$", label = "cluster1")

            ax.scatter(x_train , mu_pre[0] , c = "g" , marker = "$0$", label = "mu 0")
            ax.scatter(x_train , mu_pre[1] , c = "y" , marker = "$1$", label = "mu 1")

            ax.plot(xx , mu_pre_bound[0] , c = "b" )
            ax.plot(xx , mu_pre_bound[1] , c = "r" )

            std = []
            for i in range(num_of_cluster):
                std.append(np.sqrt(var_pre_bound[i]))
            plt.fill_between(xx , mu_pre_bound[0] - std[0] , mu_pre_bound[0] + std[0] , alpha = 0.4 , color = "g")
            plt.fill_between(xx , mu_pre_bound[1] - std[1] , mu_pre_bound[1] + std[1] , alpha = 0.4 , color = "y")
            plt.title("GPCRM_ObjectValue" + str(object_value))
            plt.ylim(-2.5 , 2.0 , 0.5)
            plt.xlim(-3.0 , 3.0 , 0.5)
            plt.legend()
#             plt.savefig("./c-ガウス過程/正則化なしパラ最適化ガウス過程c-回帰初期値100変更/params" +str(para1)+ "seed"+str(seed)+"目的関数値"+str(object_value) + ".png")
            plt.show()

            print("帰属度更新なし")
            break

def gp_ridge_regression(x_train , y_train , x_bound , kernel , params , lam):
    N = len(x_train)
    M = len(x_bound)
    K = kernel_matrix(x_train , kernel , params)
    K_inv = np.linalg.inv(K)
    K_inv_ridge = np.linalg.inv(K + lam*np.eye(N))
    mu_list = []
    var_list = []
    for x in x_bound:
        k_a=  calc_k_a(x_train , x , kernel ,params)
        k_aa = kernel(x , x , params)
        mu_list.append(k_a.T @ K_inv_ridge @ y_train)
        var_list.append(k_aa - k_a.T @ K_inv @ k_a)
    return mu_list , var_list

def gpr_ridge_plot(x_train , y_train , kernel , params , num_grid , lam):
    xx = np.linspace(np.min(x_train) , np.max(x_train) , num_grid)
    mu_pre , var_pre = gp_ridge_regression(x_train , y_train , xx , kernel , params , lam)    
    # plotーーーーーーーーーーーーーーーーーーーーーーーーーーーーーーーー
    plt.figure(figsize=(10,10))
    plt.scatter(x_train , y_train , s = 50)
    plt.plot(xx , mu_pre , c = "b")
    std = np.sqrt(var_pre)
#     print("~~~~~~~~~~~~~~~~~~~~~~~~",std)
    plt.fill_between(xx , mu_pre - std , mu_pre + std , alpha = 0.4 , color = "g")
    plt.ylim(-2.5 , 2.0 , 0.5)
    plt.xlim(-3.0 , 3.0 , 0.5)
    plt.legend()
    plt.show()
    
    
    
def gp_c_ridge_regression(x_train , y_train , x_bound , kernel , params , num_of_cluster , U , lam):
    N = len(x_train)

    result_mu_list = []
    result_var_list = []
    result_w = []
    c_params = []

    for i in range(num_of_cluster):
        u_list = []
        u_list.append([idx for idx , value in enumerate(U) if value == i])
        
        c_x = x_train[u_list[0]]
        c_y = y_train[u_list[0]]
#         params = optimize_param( c_x, c_y, kernel , init=params)
#         print("-最適化後のパラメータ-クラスタ", str(i) , params)

        K = kernel_matrix(c_x , kernel , params)
        K_inv_ridge = np.linalg.inv(K + lam*np.eye(len(c_x)))
        K_inv = np.linalg.inv(K)
        xx = np.linspace()
        mu_list = []
        var_list = []
        for x in x_bound:
            k_a=  calc_k_a(c_x , x , kernel ,params)
            k_aa = kernel(x , x , params)
            mu_list.append(k_a.T @ K_inv_ridge @ c_y)
            var_list.append(k_aa - k_a.T @ K_inv_ridge @ k_a)
        result_w.append(k_a.T @ K_inv_ridge)
        result_mu_list.append(mu_list)
        result_var_list.append(var_list)
        
    return result_mu_list , result_var_list

# def gpr_plot(x_train , y_train , kernel , params , num_grid):
def gprcrm_plot(x_train , y_train, x_bound , kernel , params , num_of_cluster , _iter , lam ,path , label, t,seed):
    
    xx = np.linspace(np.min(x_train) , np.max(x_train) , 500)
    N =  len(x_train)
    count = t
    object_value_count = 0
    old_object_value = np.inf
    ret_mu_pre = []
    ret_var_pre = []
    ret_mu_pre_bound = []
    ret_var_pre_bound = []
    ret_U = []
    ari =  0
    old_ari = 0
    t1 = time.time()
    for i in tqdm(range(50)):
        U = []
        np.random.seed(seed)
        for k in range(N):
            U.append(np.random.randint(num_of_cluster))
        hantei = False

        while hantei == False:

            mu_pre , var_pre = gp_c_ridge_regression(x_train , y_train, x_bound , kernel , params , num_of_cluster , U , lam)
            ##分散の可視化のための部分
            mu_pre_bound , var_pre_bound  = gp_c_ridge_regression(x_train , y_train, xx , kernel , params , num_of_cluster , U , lam)
            new_U = calc_U(x_train , y_train , num_of_cluster , mu_pre)
            old_U = U
            U = new_U

           # 収束条件追加
            hantei0 = np.array(old_U)
            hantei1 = np.array(U)
#             object_value = calc_objective_function(x_train, y_train, U , mu_pre , num_of_cluster, kernel , params,lam , w)
            if all(hantei1 == hantei0):
                hantei  = True
#                 ari = adjusted_rand_score(label , U)
                object_value = calc_objective_function(x_train, y_train, U , mu_pre , num_of_cluster, kernel , params,lam)

        if old_object_value > object_value:
            ret_U = U
#             old_ari = ari
            old_object_value = object_value
            ret_mu_pre = mu_pre
            ret_var_pre = var_pre
            ret_mu_pre_bound = mu_pre_bound
            ret_var_pre_bound = var_pre_bound
    t2 = time.time()
#     print("経過時間" , t2-t1)
    time_result = 0
    time_result = t2-t1
    aaa = adjusted_rand_score(label , ret_U)
    print("ARIの値" , aaa)
    
    plot_u_list = []
    for i in range(num_of_cluster):
        plot_u_list.append([idx for idx , value in enumerate(ret_U) if value == i])

    fig = plt.figure(figsize=(5,5))
    ax = fig.add_subplot(1,1,1)
    cc  = ["b", "r"  , "g"]
    mm = ["1", "2" , "3"]
    for i in  range(num_of_cluster):        
        ax.scatter(x_train[plot_u_list[i]] , y_train[plot_u_list[i]] , s = 50 , c = cc[i] , marker = "o")# , label = "cluster0")
    
    for i in range(num_of_cluster):
        #print(ret_mu_pre_bound[i])
        ax.plot(xx , ret_mu_pre_bound[i] , c = cc[i] )

    std = []
    for i in range(num_of_cluster):
        std.append(np.sqrt(np.squeeze(ret_var_pre_bound[i])))
        plt.fill_between(xx , np.squeeze(ret_mu_pre_bound[i]) - std[i] , np.squeeze(ret_mu_pre_bound[i]) + std[i] , alpha = 0.4 , color = cc[i])
    
    plt.xlim(np.min(sumX)-1,np.max(sumX)+1,1)
    plt.ylim(np.min(sumY)-1,np.max(sumY)+1,1)
    plt.xlabel("x")
    plt.ylabel("y")
    #plt.legend()
    plt.savefig(path +str(np.round(aaa , 4)) +"パラム" + str(params)+ "λ"+ str(lam) + ".png")# + "seed10state10param0.4.png")
    #plt.savefig(os.path.join(path , "ARI" +str(np.round(aaa , 4))+"経過時間" +str(np.round(time_result , 4)) + "回数"  +  str(t) + ".png"))# + "seed10state10param0.4.png")
    plt.show()
    
#     plt.close()
    return aaa , ret_U , ret_mu_pre , ret_var_pre ,time_result

            
# 目的関数計算
def calc_objective_function(X, Y, U , mu , num_of_cluster , kernel , params, lam):
    N = len(X)
#     print("###" , type(U))
    u_list = calc_label(U , num_of_cluster)
    object_value = 0
    for i in range(num_of_cluster):
        u_list_sub = []
        u_list_sub.append([idx for idx , value in enumerate(U) if value == i])
        for k in range(N):
            object_value += np.round(u_list[i][k] * ( Y[k] - mu[i][k])**2 ,4)
    return object_value

        
        
#--------------------パラメータ最適化の実装ーーーーーーーーーーーーーー
def tr(A,B):
    return (A*B.T).sum()

def likelihood(params , x_train , y_train ,kernel):
    N = len(x_train)
    K = kernel_matrix(x_train , kernel , params)
#     print("##カーネル" , np.linalg.slogdet(K))
    K_inv = np.linalg.inv(K)
    return np.linalg.slogdet(K)[1] + y_train.T @ K_inv @ y_train

def tau_grad(xi , xj , kernel , params):
    tau , sigma , eta = params
    return kernel(xi ,xj, params) - np.exp(eta) * fdelta(xi,xj)

def sigma_grad(xi , xj , kernel , params):
    tau , sigma , eta = params
    return (kernel(xi , xj , params) -  np.exp(eta) * fdelta(xi,xj)) *((xi-xj)**2 / np.exp(sigma))

def eta_grad(xi , xj , kernel , params):
    tau , sigma , eta = params
    return np.exp(eta) * fdelta(xi,xj)

def gradient(params , x_train , y_train , kernel ):
    N = len(x_train)
    K = kernel_matrix(x_train , kernel , params)
    K_inv = np.linalg.inv(K)
    K_inv_y = K_inv @ y_train
    D = len(params)
    grad = np.zeros(D)
    G1 = np.array([tau_grad(xi , xj , kernel , params) for xi in x_train for xj in x_train]).reshape(N ,N)
    grad[0] =  np.trace(K_inv @ G1) - (K_inv_y.T @ G1 @ K_inv_y)
    G2 = np.array([sigma_grad(xi , xj , kernel , params) for xi in x_train for xj in x_train]).reshape(N ,N)
    grad[1] = np.trace(K_inv @ G2) - (K_inv_y.T @ G2 @ K_inv_y)    
    G3 = np.array([eta_grad(xi , xj , kernel , params) for xi in x_train for xj in x_train]).reshape(N ,N)
    grad[2] = np.trace(K_inv @ G3) - (K_inv_y.T @ G3@ K_inv_y)

    return grad 

def optimize_fmin(x_train , y_train , kernel , init ):
    x, f, d = fmin_l_bfgs_b(func = likelihood , x0 =init , 
                            fprime=gradient ,iprint=0, maxiter=10000, 
                            args = [x_train , y_train , kernel])
    return x

def optimize (xtrain, ytrain, kernel, init):
    res = minimize (likelihood, init, args = (xtrain,ytrain,kernel),
                    jac = gradient, # numgrad
                    method = 'L-BFGS-B', #callback = print(params),
                    options = {'gtol' : 1e-2, 'disp' : True})
#     print(res.message , "パラメータ" , res.x)
    return res.x
# print("----------------------------------------")

def rbf_grad(xi , xj , kernel , param):
    theta = param
#     print(rbf_saiteki(5 , 3 , 3))# * np.exp(theta) *  (xi  - xj )**2)
    return -rbf_saiteki(xi , xj , param) * np.exp(theta) *  (xi  - xj )**2

def grad_rbf(params , x_train , y_train , kernel):
    N = len(x_train)
    K = kernel_matrix(x_train , kernel , params)
    K_inv = np.linalg.inv(K)
    K_inv_y = K_inv @ y_train
    G = np.array([rbf_grad(xi , xj , kernel , params) for xi in x_train for xj in x_train]).reshape(N ,N)
    return np.trace(K_inv @ G) - (K_inv_y.T @ G @ K_inv_y)

def optimize_param (xtrain, ytrain, kernel, init):
    res = minimize (likelihood, init, args = (xtrain,ytrain,kernel),
                    jac = rbf_grad, # numgrad
                    method = 'BFGS', #callback = print(params),
                    options = {'gtol' : 1e-2, 'disp' : True})
    print(res.message , "パラメータ" , res.x)
    return res.x
```
